```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
flight = pd.read_csv('운항 합계 데이터.csv')

cargo = pd.read_csv('화물 합계 데이터.csv')
```

```python
flight
```

```python
flight.drop('Unnamed: 0', axis=1, inplace=True)

flight 
```

```python
flight.index = ['운항 합계']

flight
```

```python
#한글깨지지 않는 코드 

from matplotlib import font_manager, rc
```

```python
plt.rcParams['axes.unicode_minus'] = False
rc('font', family='Apple SD Gothic Neo', size=20, weight='bold')

flight.T.plot(kind='bar', figsize=(len(flight.columns), 15), color='cornflowerblue')

plt.grid(True, axis='y', color='black', alpha=0.5, linestyle='-.')

plt.gca().spines['right'].set_visible(False) #오른쪽 테두리 제거
plt.gca().spines['top'].set_visible(False) #위 테두리 제거
# plt.gca().spines['left'].set_visible(False) #왼쪽 테두리 제거

plt.gca().spines['left'].set_color('blue') # y축 색상 
plt.gca().spines['bottom'].set_color('blue') #x축 색상
plt.xticks(color='navy', fontsize =30) #xticks 색상과 폰트 크기 조정
plt.yticks(color='navy', fontsize =35) #yticks 색상과 폰트 크기 조정

# plt.title('항공편 추이', fontsize=50)
plt.xlabel('[년도 / 월]', color = 'darkslategrey', fontsize=32, fontweight = 'bold', loc='right', labelpad=20)
plt.ylabel('운항  합계', color = 'darkslategrey', fontsize=32, fontweight = 'bold', labelpad=30)

plt.legend(loc='upper right', fontsize=0, frameon=False, borderpad=3) # 범례 표시 x

plt.show()
```

```python
cargo.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
cargo.index = ['화물 합계']
```

```python
plt.rcParams['axes.unicode_minus'] = False
rc('font', family='Apple SD Gothic Neo', size=20, weight='bold')

cargo.T.plot(kind='bar', figsize=(len(cargo.columns), 15), color = 'slategray')

plt.grid(True, axis='y', color='black', alpha=0.5, linestyle='-.')

plt.title('화물 추이', fontsize=25)
plt.xlabel('[년도 / 월]', fontsize=21, loc='right', labelpad=15)
plt.ylabel('화물 합계(톤)', fontsize=21, labelpad=15)

plt.legend(loc='upper right', fontsize=20, frameon=False, borderpad=3)



plt.show()
```
