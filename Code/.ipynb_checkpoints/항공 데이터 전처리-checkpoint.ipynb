```python
df = pd.read_csv('항공 데이터1.csv')

df.head(1)
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True )
df.drop('공항별(1)', axis=1, inplace=True)

df.head(1)
```

### 운항편 데이터 전처리 

```python
# 운항(편)만 가져오기 위한 코드 

# print(df.columns)
# print(len(df.columns[0]))

columns_list = []
for i in df.columns:
    if len(i) == 8:
        columns_list.append(i)

flight = df.loc[:, columns_list]
        
flight
```

```python
flight.drop([0, 1], inplace=True)

flight
```

```python
flight.index = ['운항 합계']

flight
```

```python
for i in flight.columns:
    if str(i)[6] == "0":
        flight.rename(columns = {str(i) : str(i)[2:4] + " / " +str(i)[7:8]}, inplace = True)
    else:
        flight.rename(columns = {str(i) : str(i)[2:4] + " / " +str(i)[6:8]}, inplace = True)

flight.head(1)
flight.to_csv('운항 합계 데이터.csv', mode='w')
```

### 화물(톤) 데이터 처리 

```python
columns_list2 = []
for i in df.columns:
    if len(i) == 10:
        columns_list2.append(i)

cargo = df.loc[:, columns_list2]
        
cargo
```

```python
cargo.drop([0, 1], inplace=True)
```

```python
cargo.index = ['화물 합계']
```

```python
for i in cargo.columns:
    if str(i)[6] == "0":
        cargo.rename(columns = {str(i) : str(i)[2:4] + " / " +str(i)[7:8]}, inplace = True)
    else:
        cargo.rename(columns = {str(i) : str(i)[2:4] + " / " +str(i)[6:8]}, inplace = True)

cargo.to_csv('화물 합계 데이터.csv', mode='w')
```
