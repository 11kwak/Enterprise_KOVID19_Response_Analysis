```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('covid_decided.csv')

df.head(1)
```

```python
df.drop('Unnamed: 3', axis = 1, inplace=True)
df.drop('Unnamed: 4', axis = 1, inplace=True)
df.drop('사망자 수', axis = 1, inplace=True)

df
```

```python
month = 1
for i in range(0, 12):
    df.loc[i, '날짜'] = '20 / %d' %month
    month += 1
    
df
```

```python
month = 1
for i in range(12, 18):
    df.loc[i, '날짜'] = '21 / %d' %month
    month += 1
    
df
```

```python
#한글깨지지 않는 코드 

from matplotlib import font_manager, rc
rc('font', family='Apple SD Gothic Neo', size=30, weight='bold')
```

```python
fig = plt.figure(figsize=(35, 20))
# fig.set_facecolor('white')
ax = fig.add_subplot()

ax.plot(df['날짜'], df['확진자 수'], color = 'r',linewidth = 6)


plt.xticks(rotation=90) # x축 눈금 라벨 - 90도 회전 

plt.gca().spines['right'].set_visible(False) #오른쪽 테두리 제거
plt.gca().spines['top'].set_visible(False) #위 테두리 제거
# plt.gca().spines['left'].set_visible(False) #왼쪽 테두리 제거

plt.gca().spines['left'].set_color('#00517C') #y축 색상
plt.gca().spines['bottom'].set_color('#00517C') #x축 색상
plt.xticks(color='#00517C', fontsize =30, fontweight='bold') #xticks 색상과 폰트 크기 조정
plt.yticks(color='#00517C', fontsize =35, fontweight='bold') #yticks 색상과 폰트 크기 조정

plt.ylim(0, 160000)

plt.grid(True, axis='y', color='black', alpha=1, linestyle='-.') # 가로 보조 선 

plt.fill_between(df['날짜'], df['확진자 수'], color = 'r', alpha=0.5) # 색 채우기


# plt.title('< 누적 확진자 수 >', color ='#00517C', fontsize=50, fontweight = 'bold')
plt.xlabel('[년도 / 월]', color = 'gray', fontsize=38, fontweight = 'bold', loc='right', labelpad=20)
plt.ylabel('확진자 수(명)', color = 'gray', fontsize=38, fontweight = 'bold', labelpad=30)


plt.show()
```
