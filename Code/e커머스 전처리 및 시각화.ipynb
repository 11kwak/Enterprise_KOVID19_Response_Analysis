```python
df = pd.read_csv('e커머스 매출.csv')

df.head(1)
```

```python
df.drop('Unnamed: 0', axis = 1, inplace=True)
df.drop('상품군별', axis = 1, inplace=True)
df.drop('판매매체별', axis = 1, inplace=True)
 
df.head(1)
```

```python
for i in df.columns:
    if str(i)[6] == "0":
        df.rename(columns = {str(i) : str(i)[2:4] + " / " +str(i)[7:8]}, inplace = True)
    else:
        df.rename(columns = {str(i) : str(i)[2:4] + " / " +str(i)[6:8]}, inplace = True)
        
df.head(1)
```

```python
df = df.loc[[0], :]

df.index = ['거래액']
```

```python
#한글깨지지 않는 코드 

from matplotlib import font_manager, rc


```

```python
plt.rcParams['axes.unicode_minus'] = False
rc('font', family='Apple SD Gothic Neo', size=20, weight='bold')

df.T.plot(kind='bar', figsize=(len(df.columns), 15), color='tan')
# (0.2, 0.4, 0.6)
plt.grid(True, axis='y', color='black', alpha=0.5, linestyle='-.')



plt.gca().spines['right'].set_visible(False) #오른쪽 테두리 제거
plt.gca().spines['top'].set_visible(False) #위 테두리 제거
# plt.gca().spines['left'].set_visible(False) #왼쪽 테두리 제거

plt.gca().spines['left'].set_color('brown') # y축 색상 
plt.gca().spines['bottom'].set_color('brown') #x축 색상
plt.xticks(color='brown', fontsize =30) #xticks 색상과 폰트 크기 조정
plt.yticks(color='brown', fontsize =35) #yticks 색상과 폰트 크기 조정


# plt.title('< 온라인 및 모바일 쇼핑 거래액 추이 >', color='sienna', fontsize=40)
plt.xlabel('[년도 / 월]', fontsize=32, fontweight = 'bold', loc='right', labelpad=20)
plt.ylabel('거래액 (단위: 억원)', fontsize=32, fontweight = 'bold', labelpad=30)

plt.legend(loc='upper left', fontsize=0, frameon=False, borderpad=2) # 없앰 
plt.show()
```
